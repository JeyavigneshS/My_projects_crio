## ✅ Step 1: Load the Dataset

```python
# TODO 1.1: Load the dataset using pandas and display the first few rows
import pandas as pd
import numpy as np
url="https://gitlab.crio.do/me_notebook/me_jupyter_bankattritionanalysis/-/raw/master/bank_attrition_dataset.csv"
df=pd.read_csv(url)
df.head()
```

```python
# TODO 1.2: Load the dataset using pandas by skipping row containing metadata and display the first few rows

df=pd.read_csv(url,skiprows=[1],header=0)

df.head()
```

## ✅ Step 2: Check Column Headers

```python
# TODO 2.1: Print the column names of the dataset.
df.columns
```

```python
# TODO 2.2:Standardize the column names: convert to lowercase and replace spaces with underscores.

df.columns=df.columns.str.lower().str.replace(" ","_")
df.columns
```

 ## ✅ Step 3: Handle Missing Values


```python
# TODO 3.1: Check for missing values in the dataset, Print the count of missing values in each column
df.isnull().sum()
```

```python
# TODO 3.2: Impute missing 'salary' values with median
# df['salary'] = pd.to_numeric(df['salary'], errors='coerce')  # ensure numeric
df['salary'].fillna(df['salary'].median(), inplace=True)
```

```python
# TODO 3.3: Impute missing 'balance' values with 0
df['balance'] = pd.to_numeric(df['balance'], errors='coerce')
df['balance'].fillna(0, inplace=True)
```

```python
# TODO 3.4: Impute missing 'satisfaction_score' with median
# Check for missing values in satisfaction_score
df['satisfaction_score'] = pd.to_numeric(df['satisfaction_score'], errors='coerce')
df['satisfaction_score'].fillna(df['satisfaction_score'].median(), inplace=True)
```

```python
# TODO 3.5: Drop rows where 'gender' is missing
# TODO 3.5: Drop rows where 'gender' is missing
df.dropna(subset=['gender'])
```

```python
# TODO 3.6 : Fill missing 'card_type' with the most frequent value (mode)
# Ensure the column is string
df['card_type'] = df['card_type'].astype(str)

# Replace string 'nan' with actual np.nan (assign back to avoid warning)
df['card_type'] = df['card_type'].replace('nan', np.nan)

# Fill missing values with mode using inplace=True
df['card_type'].fillna(df['card_type'].mode()[0], inplace=True)
```

```python
# TODO 3.7: Print the count of missing values in each column
df.isnull().sum()
```

## ✅ Step 4: Clean Up Text Columns

```python
# TODO 4.1: Display unique card types
df['card_type'].unique()
```

```python
# TODO 4.2: Convert 'card_type' to uppercase
df['card_type']=df['card_type'].str.upper()
```

```python
# TODO 4.3: Strip whitespace from 'card_type'
df['card_type']=df['card_type'].str.strip()
```

## ✅ Step 5: Explore Numeric Columns

## ✅ Step 6: Outlier Detection and Treatment

```python
# TODO 6.1: Import matplotlib
import matplotlib.pyplot as plt
```

```python
# TODO 6.2: Plot boxplot for 'salary'
# Chart title: Salary Distribution
plt.boxplot(df['salary'])
plt.title("Salary Distribution")
plt.show()
```

```python
# TODO 6.3: Plot boxplot for 'balance'
# Chart title: Balance Distribution 
plt.boxplot(df['balance'])
plt.title("Balance Distribution")
plt.show()
```

```python
# TODO 6.4 : Identify and count outliers in salary using IQR method
Q1=df['salary'].quantile(0.25)
Q2=df['salary'].quantile(0.50)
Q3=df['salary'].quantile(0.75)
IQR=Q3-Q1
# print(IQR)
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
outliers=df[(df['salary']<lower)|(df['salary']>upper)]
num_outliers = outliers.shape[0]
print(num_outliers)
```

```python
# TODO 6.5: Cap salary outliers at the upper bound
df['salary'] = df['salary'].apply(lambda x: upper if x > upper else (lower if x < lower else x))
```

```python
# TODO 6.6 Plot Salary distribution after handling outiers
# Chart title: Salary Distribution - Outliers Handled

plt.boxplot(df['salary'])
plt.title("Salary Distribution - Outliers Handled")
plt.show()
```
