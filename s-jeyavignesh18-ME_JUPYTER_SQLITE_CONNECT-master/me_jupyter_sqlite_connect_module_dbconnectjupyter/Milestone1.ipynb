```python
import sqlite3
import pandas as pd
import requests
```

```python
# Download the remote database file
url = "https://gitlab.crio.do/me_notebook/me_jupyter_sqlite_connect/-/raw/master/my_database.db"
response = requests.get(url)

# Save it locally
with open("my_database.db", "wb") as f:
    f.write(response.content)

# TODO 1: Connect to the database my_database.db using sqlite3
# Connect to the SQLite database file
conn = sqlite3.connect('my_database.db')
cursor = conn.cursor()
```

```python
# TODO 2: Run SQL Query to fetch records from Products Table , store in a dataframe and then print the results
# Query all records from the "products" table and load into a DataFrame
df = pd.read_sql("SELECT * FROM products;", conn)
df.head()  # Display the first few rows for verification
```

```python
# TODO 3: Store the dataframe as csv file `results.csv`
# Export the DataFrame to a CSV file called products_export.csv
df.to_csv('products_export.csv', index=False)
```

```python
# TODO 4: Close connection to the database
# It is important to close the connection after operations
conn.close()
```
