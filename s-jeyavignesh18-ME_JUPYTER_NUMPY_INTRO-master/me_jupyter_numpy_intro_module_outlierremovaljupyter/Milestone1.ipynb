# Calculate Median Using NumPy

Calculate the median of a given list or array of test scores using the NumPy library. The median is the middle value when the list is sorted, and if the list has an even number of elements, it returns the average of the two middle values.

### Implement the Function

```python
import numpy as np

def calculate_median(scores):
    """
    Calculate the median of a list or numpy array of scores using NumPy.

    Args:
        scores (list or np.ndarray): numerical values (int or float)

    Returns:
        float: median value (always float even if inputs are ints)
    """
    # Convert to numpy array (if not already)
    arr = np.array(scores)
    # Use numpy's median
    median_value = np.median(arr)
    return float(median_value)
```

### Run the Test

```python
# Test: Calculate the median of a list of test scores
scores = [90, 80, 85, 70, 95]
expected_median = 85.0  # Median of the sorted list [70, 80, 85, 90, 95] is 85

try:
    median = calculate_median(scores)
    assert median == expected_median, f"Expected {expected_median}, but got {median}"
    print('✅ Test Passed: Median is correct.')
except Exception as e:
    print(f'❌ Test Failed: {e}')
```
