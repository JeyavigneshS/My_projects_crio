# Predict Movie Rating

Estimate a missing movie rating for a user based on available ratings using basic matrix operations.

Implement two functions:

- `predict_rating(ratings, user_index, movie_index)`: Use basic NumPy operations to compute the average rating from other users who have rated the target movie.
- `predict_rating_numpy(ratings, user_index, movie_index)`: Use explicit NumPy functions to perform the same task.

Missing ratings are represented as `np.nan`. If no other users have rated the movie, return `np.nan`.

### Function Implementation: predict_rating

```python
import numpy as np

def predict_rating(ratings: np.ndarray, user_index: int, movie_index: int) -> float:
    """
    Predicts the missing rating for user_index on movie_index,
    using the average rating of *other* users who rated that movie.

    ratings: 2D NumPy array (users × movies), with some entries = np.nan
    user_index: which user we are predicting for
    movie_index: which movie’s rating to predict

    Returns:
      - a float prediction (the mean of other users’ ratings for that movie)
      - or np.nan if no other user has rated that movie
    """
    # Extract all ratings for the target movie (column)
    movie_ratings = ratings[:, movie_index]
    # Boolean mask of valid (non-nan) ratings
    valid = ~np.isnan(movie_ratings)
    # Exclude the target user
    if 0 <= user_index < movie_ratings.shape[0]:
        valid[user_index] = False

    # If no valid other ratings exist, return nan
    if not np.any(valid):
        return np.nan

    # Compute mean of those other users’ ratings
    other_ratings = movie_ratings[valid]
    prediction = np.mean(other_ratings)
    return prediction

ratings = np.array([

    [4, 5, np.nan],

    [2, 3, 5],

    [5, 5, 4]

])


user_index = 0

movie_index = 2


print(predict_rating(ratings, user_index, movie_index)) 
```

### Function Implementation: predict_rating_numpy

```python
import numpy as np

def predict_rating_numpy(ratings: np.ndarray, user_index: int, movie_index: int) -> float:
    """
    Predict the missing rating for user_index on movie_index, using NumPy operations.
    Returns the mean of the other users’ ratings for that movie, excluding NaNs and that user.
    If no other ratings exist, returns np.nan.
    """
    # Extract the column for the target movie
    movie_ratings = ratings[:, movie_index]  # shape: (n_users,)
    
    # Create mask of non-nan entries
    valid_mask = ~np.isnan(movie_ratings)
    
    # Exclude the target user
    # But first guard bounds
    if 0 <= user_index < movie_ratings.shape[0]:
        valid_mask = valid_mask & (np.arange(movie_ratings.shape[0]) != user_index)
    
    # If no valid entries, return nan
    if not np.any(valid_mask):
        return np.nan
    
    # Use the mask to filter ratings
    filtered = movie_ratings[valid_mask]
    # Compute the mean (all non-nan guaranteed)
    return np.mean(filtered)
print(predict_rating(ratings, user_index, movie_index)) 
```
