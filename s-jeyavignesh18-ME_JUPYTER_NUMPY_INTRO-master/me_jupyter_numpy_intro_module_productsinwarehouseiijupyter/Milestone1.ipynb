# Manage Warehouse Inventory Using NumPy (2D Arrays)

Efficiently manage warehouse inventory for multiple products using **NumPy 2D arrays**. The task involves updating product quantities across multiple warehouses when new shipments are received and when products are shipped out.

### Implement the Function to Update Inventory

```python
import numpy as np

def update_inventory(inventory_matrix, updates, add=True):
    # TODO
    if add:
        inventory_matrix = inventory_matrix + updates
    else:
        inventory_matrix = inventory_matrix - updates
    
    return inventory_matrix
```

### Implement the Function to Display the Inventory

```python
import numpy as np

def print_inventory_2d(matrix, header):
    """
    Prints the inventory matrix with a header and properly aligned columns.

    Parameters:
    - matrix (np.ndarray): 2D array representing the inventory.
    - header (str): Header to be printed above the inventory matrix.
    """
    locations = ['Central', 'North', 'South', 'West']
    products = ['Widget', 'Gadget', 'Gizmo']

    # Print the header
    print(header)
    print(f"{'Product':<10} |", end=" ")
    print(" | ".join([f"{loc:<10}" for loc in locations]))
    print("-" * 60)

    # Print the product rows
    for product_name, row_vals in zip(products, matrix):
        print(f"{product_name:<10} |", end=" ")
        print(" | ".join([f"{val:<10}" for val in row_vals]))
    print()
```

### Run the Test for Inventory Update

```python
import numpy as np
# Labels for reference
products = ["Widget", "Gadget", "Gizmo"]
locations = ["Central", "North", "South", "West"]

# Initial inventory matrix (3 products x 4 warehouses)
inventory_matrix = np.array([
    [10,  5,  2,  0],  # Widget
    [ 3, 15,  8,  4],  # Gadget
    [ 0,  5,  1, 10],  # Gizmo
], dtype=int)

# New shipment updates (3 products x 4 warehouses)
new_shipments_2d = np.array([
    [ 2,  1,  0,  5],  # Widget
    [ 0,  2,  1,  0],  # Gadget
    [ 4,  0,  0,  2],  # Gizmo
], dtype=int)

# Shipped-out stock updates (3 products x 4 warehouses)
shipped_out_2d = np.array([
    [ 0,  2,  1,  0],  # Widget
    [ 1,  0,  2,  2],  # Gadget
    [ 0,  1,  1,  3],  # Gizmo
], dtype=int)

# Display initial inventory
print_inventory_2d(inventory_matrix, '1) Initial Inventory (2D Array):')

# Update inventory after adding new shipments
inventory_matrix = update_inventory(inventory_matrix, new_shipments_2d, add=True)
print_inventory_2d(inventory_matrix, '2) After Adding New Shipments (2D Array):')

# Update inventory after shipping out stock
inventory_matrix = update_inventory(inventory_matrix, shipped_out_2d, add=False)
print_inventory_2d(inventory_matrix, '3) After Shipping Out Stock (2D Array):')
```

### Test the Inventory Update Function

```python
import numpy as np

def test_inventory_update():
    # Test case for inventory update
    inventory_matrix = np.array([
        [10,  5,  2,  0],  # Widget
        [ 3, 15,  8,  4],  # Gadget
        [ 0,  5,  1, 10],  # Gizmo
    ], dtype=int)

    new_shipments_2d = np.array([
        [ 2,  1,  0,  5],  # Widget
        [ 0,  2,  1,  0],  # Gadget
        [ 4,  0,  0,  2],  # Gizmo
    ], dtype=int)

    # Expected updated inventory after adding shipments
    expected_inventory = np.array([
        [12,  6,  2,  5],  # Widget
        [ 3, 17,  9,  4],  # Gadget
        [ 4,  5,  1, 12],  # Gizmo
    ], dtype=int)

    # Update inventory with new shipments
    updated_inventory = update_inventory(inventory_matrix, new_shipments_2d, add=True)

    # Assert the inventory matches the expected updated inventory
    assert np.array_equal(updated_inventory, expected_inventory), f'Expected {expected_inventory}, but got {updated_inventory}'

    print('✅ Test Passed: Inventory updated correctly after adding new shipments.')

try:
    test_inventory_update()
except Exception as e:
    print(f'❌ Test Failed: {e}')
```
