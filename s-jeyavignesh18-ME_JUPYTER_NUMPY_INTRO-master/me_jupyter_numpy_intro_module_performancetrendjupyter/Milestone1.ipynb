# Calculate Mean Using NumPy

Calculate the mean (average) of a list of scores using NumPy. The mean is computed using the `np.mean()` function to provide an efficient way of determining the average of the input scores.

### Implement the Function

```python
import numpy as np

def calculate_mean(scores):
    # TODO
    scores=np.array(scores,dtype=float)
    return np.mean(scores)
```

### Run the Test

```python
# Test: Calculate mean of a list of test scores
scores = [90, 80, 85, 70, 95]
expected_mean = 84.0  # The average of [90, 80, 85, 70, 95] is 84.0

try:
    mean_result = calculate_mean(scores)
    assert mean_result == expected_mean, f"Expected {expected_mean}, but got {mean_result}"
    print('✅ Test Passed: Mean calculated correctly.')
except Exception as e:
    print(f'❌ Test Failed: {e}')
```
