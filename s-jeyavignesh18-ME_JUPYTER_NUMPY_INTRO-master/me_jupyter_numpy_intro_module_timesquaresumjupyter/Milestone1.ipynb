# Sum of Squares Calculation with Time Measurement

calculate the sum of squares of all integers from 1 to `n`:
- **List-based approach**: Using a list of integers and a for-loop to compute the sum of squares.
- **NumPy-based approach**: Using NumPy's vectorized operations for a more efficient computation.

Both approaches measure the time taken for the operation and print the results.

### Implement the List-based Sum of Squares Function

```python
import time

def sum_of_squares_list(n):
    """
    Computes the sum of squares of integers from 1 to n using a list-based approach.
    """
    start_time = time.time()
    
    # Create a list of squares
    squares = [i**2 for i in range(1, n + 1)]
    
    # Sum the squares
    total_sum = sum(squares)
    
    end_time = time.time()
    time_taken = end_time - start_time
    
    print(f"Time taken for list-based approach: {time_taken} seconds")
    return total_sum

sum_of_squares_list(4)
```
