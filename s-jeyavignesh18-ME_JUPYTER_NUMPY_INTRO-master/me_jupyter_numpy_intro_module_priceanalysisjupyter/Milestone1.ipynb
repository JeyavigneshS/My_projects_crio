# Price Analysis with NaN Exclusion

Implement the function `analyze_prices(prices)` to compute the mean, median, and mode for a list of prices (excluding any NaN values). If all values are NaN, return `None` for each metric.

### Function Implementation

```python
import numpy as np
from scipy import stats

def analyze_prices(prices):
    """
    Compute the mean, median, and mode for a list of prices, excluding NaN values.
    
    Args:
        prices (list): List of product prices (may include np.nan).
    
    Returns:
        dict: Dictionary with keys 'mean', 'median', and 'mode'.
              If all values are NaN, return None for each metric.
    """
    # Convert the list to a NumPy array
    prices_array = np.array(prices)
    
    # Remove NaN values
    valid_prices = prices_array[~np.isnan(prices_array)]
    
    # Check if there are any valid prices
    if valid_prices.size == 0:
        return {'mean': None, 'median': None, 'mode': None}
    
    # Calculate mean and median
    mean = np.mean(valid_prices)
    median = np.median(valid_prices)
    
    # Mode calculation
    mode_result = stats.mode(valid_prices, nan_policy='omit')
    mode = mode_result.mode[0] if isinstance(mode_result.mode, np.ndarray) else mode_result.mode
    
    return {'mean': mean, 'median': median, 'mode': mode}

prices = [100, 150, 200, np.nan, 250, 150, np.nan, 300]
analyze_prices(prices)
```

### Testing the Function

```python
import numpy as np

# Test Sample Input 1
prices1 = [100, 150, 200, np.nan, 250, 150, np.nan, 300]
expected_output1 = {
    "mean": 191.66666666666666,  # Correct mean: (100 + 150 + 200 + 250 + 150 + 300) / 6
    "median": 175.0,
    "mode": 150.0
}

output1 = analyze_prices(prices1)
print("Output for Sample Input 1:", output1)

# Validate the results for Sample Input 1
assert np.isclose(output1["mean"], expected_output1["mean"]), f"Expected mean {expected_output1['mean']}, got {output1['mean']}"
assert np.isclose(output1["median"], expected_output1["median"]), f"Expected median {expected_output1['median']}, got {output1['median']}"
assert output1["mode"] == expected_output1["mode"], f"Expected mode {expected_output1['mode']}, got {output1['mode']}"
print("✅ Test Passed for Sample Input 1")

# Test Sample Input 2 (all values are NaN)
prices2 = [np.nan, np.nan, np.nan]
expected_output2 = {
    "mean": None,
    "median": None,
    "mode": None
}

output2 = analyze_prices(prices2)
print("\nOutput for Sample Input 2:", output2)
assert output2 == expected_output2, f"Expected {expected_output2}, got {output2}"
print("✅ Test Passed for Sample Input 2")
```
