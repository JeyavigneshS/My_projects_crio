### Implement the `update_inventory` Function

This function should update the inventory for all four warehouses by applying the provided updates. Replace the `pass` statement with your implementation.

```python
def update_inventory(central_warehouse, north_warehouse, south_warehouse, west_warehouse, updates, add=True):
    def update_warehouse(warehouse, update):
        if add:
            return [w + u for w, u in zip(warehouse, update)]
        else:
            return [w - u for w, u in zip(warehouse, update)]
    
    central_updated = update_warehouse(central_warehouse, updates[0])
    north_updated = update_warehouse(north_warehouse, updates[1])
    south_updated = update_warehouse(south_warehouse, updates[2])
    west_updated = update_warehouse(west_warehouse, updates[3])
    
    return central_updated, north_updated, south_updated, west_updated

def print_inventory(central_warehouse, north_warehouse, south_warehouse, west_warehouse, header):
    print(header)
    print(f"{'Product':8} {'Central':>8} {'North':>8} {'South':>8} {'West':>8}")
    for i in range(len(central_warehouse)):
        print(f"{i+1:<8} {central_warehouse[i]:>8} {north_warehouse[i]:>8} {south_warehouse[i]:>8} {west_warehouse[i]:>8}")
```

### Helper Function: `print_inventory`

The following function is already implemented to display the inventory in a formatted manner.

```python
def print_inventory(central_warehouse, north_warehouse, south_warehouse, west_warehouse, header, products):
    """
    Display the inventory for all products in the four warehouses.
    
    Parameters:
      central_warehouse, north_warehouse, south_warehouse, west_warehouse: Lists of product quantities per warehouse.
      header: A string header to print before displaying the inventory.
      products: A list of product names.
    """
    print(header)
    for i in range(len(products)):
        print(f"{products[i]:<6} | Central: {central_warehouse[i]} | North: {north_warehouse[i]} | South: {south_warehouse[i]} | West: {west_warehouse[i]}")
    print()
```

### Test Cases for Inventory Update - Addition (New Shipments)

Run the following cell to test the inventory update when new shipments are received.

```python
# Define products for display
products = ["Widget", "Gadget", "Gizmo"]

# Initial warehouse quantities
central_warehouse = [10, 3, 0]
north_warehouse   = [5, 15, 5]
south_warehouse   = [2, 8, 1]
west_warehouse    = [0, 4, 10]

# New shipments to add
new_shipments_central = [2, 0, 4]
new_shipments_north   = [1, 2, 0]
new_shipments_south   = [0, 1, 0]
new_shipments_west    = [5, 0, 2]

new_shipments = [new_shipments_central, new_shipments_north, new_shipments_south, new_shipments_west]

# Update inventory by adding new shipments
central_warehouse, north_warehouse, south_warehouse, west_warehouse = update_inventory(
    central_warehouse, north_warehouse, south_warehouse, west_warehouse, new_shipments, add=True
)

# Expected inventory after addition
expected_central = [12, 3, 4]
expected_north   = [6, 17, 5]
expected_south   = [2, 9, 1]
expected_west    = [5, 4, 12]

# Display the updated inventory
print_inventory(central_warehouse, north_warehouse, south_warehouse, west_warehouse, "Inventory after adding new shipments:", products)

# Simple assertions to validate the update
assert central_warehouse == expected_central, f"Expected {expected_central}, but got {central_warehouse}"
assert north_warehouse   == expected_north,   f"Expected {expected_north}, but got {north_warehouse}"
assert south_warehouse   == expected_south,   f"Expected {expected_south}, but got {south_warehouse}"
assert west_warehouse    == expected_west,    f"Expected {expected_west}, but got {west_warehouse}"

print("✅ Test Passed: Inventory updated correctly after adding new shipments.")
```
