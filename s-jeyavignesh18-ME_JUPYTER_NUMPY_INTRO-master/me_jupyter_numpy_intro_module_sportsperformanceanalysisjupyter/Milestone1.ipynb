# Evaluate Player Performance

Implement the function `find_best_player(players)` to compute the overall performance score for each player (ignoring `np.nan` values) and return the index of the best-performing player. In case of a tie, return the lowest index.

```python
import numpy as np

def find_best_player(players: np.ndarray) -> int:
    """
    Finds the index of the best-performing player based on the sum of their skill metrics,
    ignoring any NaN values. In case of a tie, returns the lowest index.

    Parameters:
    - players: A 2D NumPy array where each row represents a player and each column represents a skill metric.

    Returns:
    - int: The index of the best-performing player.
    """
    # Compute the sum of each player's skill metrics, ignoring NaN values
    performance_scores = np.nansum(players, axis=1)

    # Return the index of the player with the highest performance score
    # In case of a tie, np.argmax returns the first occurrence
    return np.argmax(performance_scores)

players = np.array([[8, 9, 7], [6, 7, 8], [9, 8, 9]])

print(find_best_player(players))
```

### Testing

```python
def run_test_case(test_name, players, expected):
    result = find_best_player(players)
    assert result == expected, f"{test_name} Failed: Expected {expected}, got {result}"
    print(f"✅ {test_name} Passed")

# Test cases
players1 = np.array([[8, 9, 7], [6, 7, 8], [9, 8, 9]])
run_test_case("Test Case 1", players1, 2)

players2 = np.array([[5, 6, np.nan], [7, 8, 9], [5, 5, 5]])
run_test_case("Test Case 2", players2, 1)
```
