# Find the Busiest Intersection

Find the busiest intersection based on traffic data. 
We will compute the total traffic for each intersection and return the index of the one with the highest traffic.

### Implement the Function Using NumPy

```python
import numpy as np

def find_busiest_intersection_numpy(traffic_data: np.ndarray) -> int:
    total_per_intersection = np.sum(traffic_data, axis=1)
    busiest_index = int(np.argmax(total_per_intersection))
    return busiest_index

traffic_data = np.array([

    [200, 180, 220],  

    [150, 170, 190],  

    [300, 310, 290]  

])

find_busiest_intersection_numpy(traffic_data)
```

### Implement the Function Without Using np.argmax

```python
import numpy as np

def find_busiest_intersection_list(traffic_data):
    # If it's a numpy array, convert to list of lists
    if isinstance(traffic_data, np.ndarray):
        if traffic_data.size == 0:
            return None
        traffic_data = traffic_data.tolist()

    if len(traffic_data) == 0:
        return None

    totals = []
    for row in traffic_data:
        try:
            s = sum(row)
        except TypeError:
            s = row
        totals.append(s)

    max_total = max(totals)
    busiest_index = totals.index(max_total)
    return busiest_index
find_busiest_intersection_list(traffic_data)
```

### Run the Test

```python
traffic_data = np.array([
    [200, 180, 220],  
    [150, 170, 190],  
    [300, 310, 290]  
])

expected_output = 2  # Intersection 2 has the highest total traffic

try:
    # Test with NumPy method
    result_numpy = find_busiest_intersection_numpy(traffic_data)
    assert result_numpy == expected_output, f"Expected {expected_output}, but got {result_numpy}"
    
    # Test with list-based method
    result_list = find_busiest_intersection_list(traffic_data)
    assert result_list == expected_output, f"Expected {expected_output}, but got {result_list}"
    
    print('✅ Test Passed: Busiest intersection found correctly.')
except Exception as e:
    print(f'❌ Test Failed: {e}')
```
