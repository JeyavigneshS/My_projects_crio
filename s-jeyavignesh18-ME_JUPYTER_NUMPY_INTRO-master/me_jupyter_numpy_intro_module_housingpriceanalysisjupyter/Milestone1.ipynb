# Housing Price Analysis (HOUSING_PRICES)

Determine the most appropriate measure of central tendency (mean, median, or mode) for a given list of home prices. The program detects outliers and chooses the best metric accordingly.

### Implement the Function to Detect Outliers

```python
import numpy as np
from scipy import stats

def has_outliers(prices):
    return any(p > 1000000 for p in prices)
has_outliers(prices)
```

### Implement the Function to Calculate Central Tendency

```python

def find_central_tendency(prices):
    prices = np.array(prices)
    
    # Check if all values identical
    if np.all(prices == prices[0]):
        return f"Mode: {prices[0]}"
    
    # Check outliers
    if has_outliers(prices):
        median = int(np.median(prices))
        return f"Median: {median}"
    
    # No outliers, compute mean
    mean = np.mean(prices)
    
    # Check mode: if all unique, mode returns all values, no unique mode
    mode_res = stats.mode(prices, nan_policy='omit')
    mode_vals = mode_res.mode
    count = mode_res.count
    
    if len(mode_vals) == len(prices):  # all unique, no mode
        mode = None
    else:
        mode = mode_vals[0]
    
    if mode is not None:
        # This means there's a mode if not all unique, but only if all equal handled above.
        # For this problem, only return mode if all identical, which is already handled.
        pass
    
    # Return mean
    mean_int = int(mean) if mean.is_integer() else mean
    return f"Mean: {mean_int}"
find_central_tendency(prices)
```
