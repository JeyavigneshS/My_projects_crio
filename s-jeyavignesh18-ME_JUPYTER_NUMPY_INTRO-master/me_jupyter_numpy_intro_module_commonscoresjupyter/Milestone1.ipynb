# Calculate Mode Using NumPy and SciPy

Calculate the mode of a given list of test scores using NumPy and SciPy. The mode is the most frequently occurring value(s) in the data. If there are multiple values with the same highest frequency, the mode will return all of them.

### Implement the Function

```python
import numpy as np
from scipy import stats

def calculate_mode(scores):
    """
    Calculate the mode(s) of scores using SciPy + NumPy.
    
    Args:
        scores: list or np.ndarray of numeric values

    Returns:
        If there is a single mode: return that value (int or float).
        If there are multiple modes (tie): return a sorted list of all modes.
    """
    # Convert to numpy array
    arr = np.array(scores)
    
    # Edge case: empty
    if arr.size == 0:
        raise ValueError("scores must contain at least one value")
    
    # Use numpy.unique to count occurrences
    values, counts = np.unique(arr, return_counts=True)
    
    # Find the maximum count
    max_count = np.max(counts)
    
    # Find all values whose count equals max_count
    modes = values[counts == max_count]
    
    # If only one mode, return it. Else, return sorted list.
    if modes.size == 1:
        # Return scalar, but consistent type
        mode_val = modes[0]
        # If it's integral (no fractional part), you might want int
        # but depends on what you want. Here I'll return the same type as input
        return mode_val
    else:
        # multiple modes; ensure ascending order
        sorted_modes = np.sort(modes)
        return sorted_modes.tolist()
```

### Test the Function

```python
def test_calculate_mode():
    # Test 1: Regular case with one mode
    scores = [90, 80, 85, 70, 95, 85]
    expected_mode = 85

    try:
        mode = calculate_mode(scores)
        assert mode == expected_mode, f"Expected {expected_mode}, but got {mode}"
        print('✅ Test Passed: Mode is correct.')
    except Exception as e:
        print(f'❌ Test Failed: {e}')

test_calculate_mode()
```
