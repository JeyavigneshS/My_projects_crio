# Apply Salary Increase

Increase salaries by a given scalar multiplier. This function helps adjust employee salaries based on a percentage increase.

### Implement the Function

```python
import numpy as np

def apply_salary_increase(salary_matrix, scalar):
    """
    Applies a salary increase to all employees across different departments.

    Parameters:
    - salary_matrix (np.ndarray): A 2D NumPy array representing current salaries.
    - scalar (float): A scalar value representing the percentage increase (e.g., 1.10 for a 10% increase).

    Returns:
    - np.ndarray: A new matrix with updated salaries, rounded to two decimal places.
    """
    # Perform scalar multiplication and round the result
    updated_salaries = np.round(salary_matrix * scalar, 2)
    return updated_salaries


salary_matrix = np.array([

[40, 45, 50],

[55, 60, 50],

[70, 75, 80]

])


scalar = 1.10
apply_salary_increase(salary_matrix, scalar)
```

### Run the Test

```python
# Test: Applying salary increase
salary_matrix = np.array([
    [40, 45, 50],
    [55, 60, 50],
    [70, 75, 80]
])

scalar = 1.10

expected_salary_output = np.array([
    [44.0, 49.5, 55.0],
    [60.5, 66.0, 55.0],
    [77.0, 82.5, 88.0]
])

try:
    salary_output = apply_salary_increase(salary_matrix, scalar)
    assert np.array_equal(salary_output, expected_salary_output), f"Expected:\n{expected_salary_output}\nBut got:\n{salary_output}"
    print('✅ Test Passed: Salary increase function works correctly.')
except Exception as e:
    print(f'❌ Test Failed: {e}')
```
