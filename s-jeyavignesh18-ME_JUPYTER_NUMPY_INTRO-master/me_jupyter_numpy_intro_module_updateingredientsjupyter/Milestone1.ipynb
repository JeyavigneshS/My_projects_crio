# Modify Order Based on Ingredients

Modify the first ingredient of a given order. If the first ingredient is not 'orange', it will be replaced with 'orange'. The original order will remain unchanged.

### Implement the Function

```python
import copy
```

```python

def modify_order(order):
    # Keep original unchanged
    original_order = copy.deepcopy(order)
    
    # Make modified copy
    modified_order = copy.deepcopy(order)
    if modified_order and modified_order[0] != 'orange':
        modified_order[0] = 'orange'
    return modified_order,original_order
```

### Run the Test

```python
# Test: Replacing 'apple' with 'orange' in the first dish
order = ['apple', 'banana', 'carrot']
expected_modified = ['orange', 'banana', 'carrot']  # 'apple' should be replaced with 'orange'
expected_original = ['apple', 'banana', 'carrot']

try:
    modified, original = modify_order(order)
    assert modified == expected_modified, f"Expected {expected_modified}, but got {modified}"
    assert original == expected_original, f"Expected {expected_original}, but got {original}"
    print("**✅ Test Passed:** First ingredient replaced successfully when necessary.")
except Exception as e:
    print(f"**❌ Test Failed:** {e}")
```
