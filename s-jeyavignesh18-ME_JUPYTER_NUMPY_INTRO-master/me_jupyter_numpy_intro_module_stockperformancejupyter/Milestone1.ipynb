# Portfolio Return Calculation

Calculate the overall return of an investment portfolio, where each stock has a return percentage and a corresponding weight in the portfolio.

### Implement the Function

```python
import numpy as np

def portfolio_return(returns, weights):
    """
    Computes the overall portfolio return using the dot product.

    Parameters:
    - returns (list): List of return percentages for each stock.
    - weights (list): List of weight percentages for each stock (should sum to 100).

    Returns:
    - float: The overall portfolio return as a weighted average.
    """
    # Convert weights to decimal form
    weights_decimal = np.array(weights) / 100

    # Compute the dot product of returns and weights
    weighted_return = np.dot(returns, weights_decimal)

    return weighted_return

returns = [10, 5, 8]

weights = [50, 30, 20]

print(portfolio_return(returns, weights))
```

### Run the Test

```python
# Test: Calculating portfolio return
returns = [10, 5, 8]
weights = [50, 30, 20]
expected_output = 8.1

try:
    output = portfolio_return(returns, weights)
    assert round(output, 2) == round(expected_output, 2), f"Expected {expected_output}, but got {output}"
    print("✅ Test Passed: Portfolio return function works correctly.")
except Exception as e:
    print(f"❌ Test Failed: {e}")
```
