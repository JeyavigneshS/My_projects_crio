# Compare Memory Usage of Python List and NumPy Array

Compare the memory usage of a Python list and a NumPy array containing integers from `0` to `n-1`. The comparison is done using `sys.getsizeof()` for the Python list and `.nbytes` for the NumPy array.

### Implement `memory_usage_list` Function

```python
import sys
import numpy as np

def memory_usage_list(n):
    lst = list(range(n))
    container_memory = sys.getsizeof(lst)
    elements_memory = sum(sys.getsizeof(elem) for elem in lst)
    total_memory = container_memory + elements_memory
    print(f"Python list container only: {container_memory} bytes")
    print(f"Python list (container + elements): {total_memory} bytes")
    return total_memory

```

### Implement `memory_usage_numpy` Function

```python
def memory_usage_numpy(n):
    arr = np.arange(n, dtype=int)
    array_memory = arr.nbytes
    print(f"NumPy array: {array_memory} bytes")
    return array_memory
```

### Test the Functions

```python
# Test case: Compare memory usage for n=5
print('Testing for n=5:')
try:
    list_memory = memory_usage_list(5)
    numpy_memory = memory_usage_numpy(5)
    print(f'✅ Test Passed: List and NumPy memory comparison completed for n=5.')
except Exception as e:
    print(f'❌ Test Failed: {e}')

```
