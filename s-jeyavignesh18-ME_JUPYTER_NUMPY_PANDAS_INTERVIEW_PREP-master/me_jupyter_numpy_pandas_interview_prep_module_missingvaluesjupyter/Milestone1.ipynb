```python
def clean_sales_data(dataframe):
    # Fill missing Revenue values with mean of non-missing values
    dataframe["Revenue"] = dataframe["Revenue"].fillna(dataframe["Revenue"].mean(skipna=True))

    # Replace "Unknown" in Region with "Not Specified"
    dataframe["Region"] = dataframe["Region"].replace("Unknown", "Not Specified")

    # Drop rows where Product is missing
    dataframe = dataframe.dropna(subset=["Product"])

    return dataframe
```

```python
# Do not edit this cell
import pandas as pd
import inspect

def test_clean_sales_data():
    """Test the clean_sales_data function using the actual sales dataset."""

    # Load dataset from URL
    url = "https://gitlab.crio.do/root/me_jupyter_numpy_pandas_interview_prep_datasets/-/raw/master/sales_performance.csv"
    original_df = pd.read_csv(url)

    # Make a copy for testing
    test_df = original_df.copy()

    # Get source code to validate method usage
    source_code = inspect.getsource(clean_sales_data)

    # Calculate expected mean revenue (before cleaning)
    expected_mean = test_df['Revenue'].mean(skipna=True)

    # Count rows with missing Product (to check if they're dropped)
    rows_with_missing_product = test_df['Product'].isnull().sum()
    expected_row_count = len(test_df) - rows_with_missing_product

    # Run the cleaning function
    cleaned_df = clean_sales_data(test_df)

    # Assertions
    assert isinstance(cleaned_df, pd.DataFrame), "Function should return a DataFrame"
    assert "fillna" in source_code, "fillna not used to handle missing Revenue"
    assert "replace" in source_code, "replace not used to handle Region values"
    assert "dropna" in source_code, "dropna not used to drop rows with missing Product"

    # Check if missing Revenue is filled with expected mean
    assert cleaned_df['Revenue'].isnull().sum() == 0, "There are still missing values in Revenue"
    assert round(expected_mean, 2) in cleaned_df['Revenue'].round(2).values, "Mean Revenue not used to fill missing values"

    # Check if 'Unknown' in Region replaced
    assert 'Unknown' not in cleaned_df['Region'].values, "'Unknown' was not replaced in Region"
    assert 'Not Specified' in cleaned_df['Region'].values, "'Not Specified' not found in Region"

    # Check rows with missing Product are dropped
    assert cleaned_df['Product'].isnull().sum() == 0, "Rows with missing Product were not dropped"
    assert len(cleaned_df) == expected_row_count, "Incorrect number of rows after dropping missing Product"

    print("✅ Test Passed: clean_sales_data() works correctly on the provided dataset.")

# Run the test
test_clean_sales_data()
```
