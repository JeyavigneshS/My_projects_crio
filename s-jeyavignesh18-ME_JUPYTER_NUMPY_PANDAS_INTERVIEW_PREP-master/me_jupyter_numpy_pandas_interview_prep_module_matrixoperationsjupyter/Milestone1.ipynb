```python
import numpy as np
import pandas as pd

# URL for the dataset
url = "https://gitlab.crio.do/root/me_jupyter_numpy_pandas_interview_prep_datasets/-/raw/master/sales_performance.csv"

def analyze_sales_data(data: np.ndarray) -> dict:
    """
    data: 2D NumPy array, shape (months, products)
    Returns a dict with keys:
      - "total_revenue": np.ndarray (sum per product)
      - "mean_revenue": float (overall mean)
      - "discounted_sales": np.ndarray (same shape, after discount)
      - "highest_average_product_index": int (index of best product by post-discount mean)
    """
    # Compute total revenue per product (column sums, ignoring NaNs)
    total_revenue = np.nansum(data, axis=0)
    
    # Compute overall mean across all entries (ignoring NaNs)
    mean_revenue = float(np.nanmean(data))
    
    # Create discounted sales array (float, same shape)
    discounted_sales = data.astype(float).copy()
    mask = discounted_sales > 5000
    discounted_sales[mask] = discounted_sales[mask] * 0.9
    
    # Compute mean per product after discount
    mean_per_product_after = np.nanmean(discounted_sales, axis=0)
    
    # Find index of maximum
    highest_idx = int(np.nanargmax(mean_per_product_after))
    
    return {
        "total_revenue": total_revenue,
        "mean_revenue": mean_revenue,
        "discounted_sales": discounted_sales,
        "highest_average_product_index": highest_idx
    }

# Test harness integration (with your test function)
def test_analyze_sales_data():
    """Test the analyze_sales_data function with actual dataset."""
    # Load dataset via pandas
    df = pd.read_csv(url)
    # Select numeric columns (floats, ints) only
    sales_data = df.select_dtypes(include=[float, int]).to_numpy()
    # Run analysis
    result = analyze_sales_data(sales_data)
    # Extract outputs
    total_revenue = result["total_revenue"]
    mean_revenue = result["mean_revenue"]
    discounted_sales = result["discounted_sales"]
    highest_avg_index = result["highest_average_product_index"]
    
    # Checks
    assert isinstance(total_revenue, np.ndarray), "total_revenue should be a NumPy array"
    assert isinstance(mean_revenue, float) or isinstance(mean_revenue, np.floating), "mean_revenue should be a float"
    assert isinstance(discounted_sales, np.ndarray), "discounted_sales should be a NumPy array"
    assert isinstance(highest_avg_index, int), "highest_average_product_index should be an integer"
    
    # No NaNs in total_revenue or mean_revenue
    assert not np.isnan(total_revenue).any(), "total_revenue should not contain NaN"
    assert not np.isnan(mean_revenue), "mean_revenue should not be NaN"
    
    # Discount correctness
    # For entries ≤ 5000, discounted should equal original
    assert np.all(discounted_sales[sales_data <= 5000] == sales_data[sales_data <= 5000]), "Values ≤ 5000 should not be discounted"
    # For entries > 5000, discounted should be 90% of original
    assert np.all(discounted_sales[sales_data > 5000] == sales_data[sales_data > 5000] * 0.9), "Values > 5000 should be discounted"
    
    # Index bounds
    assert 0 <= highest_avg_index < sales_data.shape[1], "highest_average_product_index out of bounds"
    
    print("✅ Test Passed: analyze_sales_data() works correctly on the provided dataset.")

# Run the test
if __name__ == "__main__":
    test_analyze_sales_data()
```
