```python
import pandas as pd

def load_btc_market_price_with_parsed_dates():
    """Load BTC market price data, ensuring the Timestamp column is parsed as datetime."""
    url = "https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/btc-market-price.csv"
    
    # Define the date format (adjust as per your data)
    date_format = "%d/%m/%Y %H:%M:%S"  # Example: '24/09/2025 12:52:19'
    
    # Custom date parser function
    def date_parser(x):
        return pd.to_datetime(x, format=date_format)
    
    # Load CSV with specified date parser
    df = pd.read_csv(url, header=None, names=['Timestamp', 'MarketPrice'], parse_dates=[0], date_parser=date_parser)
    
    return df
load_btc_market_price_with_parsed_dates()
```
