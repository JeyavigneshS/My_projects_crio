```python
import pandas as pd

# Initialize the DataFrame
df = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [1000, 25, 50, 300],
    'Quantity': [10, 100, 50, 30]
})

# Define the function to calculate total value
def calculate_total_value(df):
    """
    Calculates the total value of each product.

    Args:
        df (pd.DataFrame): DataFrame containing product data.

    Returns:
        pd.DataFrame: DataFrame with an additional 'Total Value' column.
    """
    df['Total Value'] = df['Price'] * df['Quantity']
    return df

# Call the function
df = calculate_total_value(df)

# Display the resulting DataFrame
print(df)
```

```python
# Do not edit this cell
def test_calculate_total_value():
    df = create_product_dataframe()
    df = calculate_total_value(df)
    assert "Total Value" in df.columns, "Total Value column missing"
    assert all(df["Total Value"] == df["Price"] * df["Quantity"]), "Total Value calculation incorrect"
    print("\n✅ Test Passed: Total Value calculated successfully.")
    print(df)

# Run the tests
test_calculate_total_value()
```
