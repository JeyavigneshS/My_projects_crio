```python
import pandas as pd

def create_monthly_sales_series():
    """
    Creates a Pandas Series with monthly sales data.

    Returns:
        pd.Series: Series containing sales data indexed by month.
    """
    # Define the months and corresponding sales figures
    months = ["January", "February", "March", "April", "May", "June"]
    sales = [200, 300, 250, 400, 350, 500]
    
    # Create and return the Pandas Series
    return pd.Series(sales, index=months)

[200, 300, 250, 400, 350, 500]

["January", "February", "March", "April", "May", "June"]


create_monthly_sales_series()
```

```python
# Do not edit this cell
def test_create_monthly_sales_series():
    sales = create_monthly_sales_series()
    assert not sales.empty, "Sales series is empty"
    assert list(sales.index) == ["January", "February", "March", "April", "May", "June"], "Month index mismatch"
    assert list(sales.values) == [200, 300, 250, 400, 350, 500], "Sales values mismatch"

    print("\n✅ Test Passed: Monthly Sales Series created successfully.")
    print(sales)

# Run the tests
test_create_monthly_sales_series()
```

```python
import pandas as pd

def get_sales_for_month(sales, month):
    """
    Retrieves the sales data for a specific month.

    Args:
        sales (pd.Series): Series containing monthly sales data indexed by month names.
        month (str): The month for which sales data is to be retrieved.

    Returns:
        int: Sales value for the given month.
    """
    return sales.get(month, "Month not found")

# Define sales and months
sales = [200, 300, 250, 400, 350, 500]
months = ["January", "February", "March", "April", "May", "June"]

# Convert lists to Pandas Series
sales_series = pd.Series(sales, index=months)

# Retrieve sales for March
march_sales = get_sales_for_month(sales_series, "March")
print(f"Sales in March: {march_sales}")
```

```python
def test_get_sales_for_month():
    # Create the monthly sales series
    sales = create_monthly_sales_series()
    
    # Specify the month you're interested in
    month = "March"
    
    # Retrieve sales for the specified month
    march_sales = get_sales_for_month(sales, month)
    
    # Assert that the sales value for March is correct
    assert march_sales == 250, f"Expected 250, but got {march_sales}"

    print("\n✅ Test Passed: Sales for March retrieved successfully.")
    print(f"Sales in March: {march_sales}")

# Run the test
test_get_sales_for_month()
```

```python
import pandas as pd

def update_sales_for_month(sales, month, new_sales_value):
    """
    Updates the sales data for a specific month.

    Args:
        sales (pd.Series): Series containing monthly sales data indexed by month names.
        month (str): The month for which sales data is to be updated.
        new_sales_value (int): The new sales value to set for the specified month.

    Returns:
        pd.Series: Updated sales series.
    """
    if month in sales.index:
        sales[month] = new_sales_value
    else:
        print(f"Month '{month}' not found in the sales data.")
    return sales

# Sample monthly sales data
sales_data = {
    "January": 200,
    "February": 300,
    "March": 250,
    "April": 400,
    "May": 350,
    "June": 500
}

# Create a Pandas Series
sales_series = pd.Series(sales_data)

# Define the new sales value for June
new_sales_value = 550

# Update sales for June
updated_sales = update_sales_for_month(sales_series, "June", new_sales_value)
print(updated_sales)
```

```python
def test_update_sales_for_month():
    # Create the monthly sales series
    sales = create_monthly_sales_series()
    
    # Specify the month and the new sales value
    month = "June"
    new_sales_value = 550
    
    # Update sales for the specified month
    updated_sales = update_sales_for_month(sales, month, new_sales_value)
    
    # Assert that the sales value for June is updated correctly
    assert updated_sales["June"] == 550, "June sales update failed"

    print("\n✅ Test Passed: Sales for June updated successfully.")
    print(updated_sales)

# Run the test
test_update_sales_for_month()
```

```python
import pandas as pd

def increase_sales_by_percentage(sales, percentage):
    """
    Increases each sales figure in the series by the specified percentage.

    Args:
        sales (pd.Series): Series containing monthly sales data.
        percentage (float): The percentage by which to increase each sales figure.

    Returns:
        pd.Series: Updated sales series with increased values.
    """
    return sales * (1 + percentage / 100)

# Sample monthly sales data
sales_data = {
    "January": 200,
    "February": 300,
    "March": 250,
    "April": 400,
    "May": 350,
    "June": 500
}

# Create a Pandas Series
sales_series = pd.Series(sales_data)

# Define the percentage increase
percentage = 10

# Increase sales by the specified percentage
updated_sales = increase_sales_by_percentage(sales_series, percentage)
print(updated_sales)
```

```python
def test_increase_sales_by_percentage():
    sales = create_monthly_sales_series()
    percentage = 10  # Define the percentage increase
    increased_sales = increase_sales_by_percentage(sales, percentage)
    expected_values = [value * 1.10 for value in sales.values]
    assert all(increased_sales.values == expected_values), "Sales increase calculation incorrect"

    print("\n✅ Test Passed: Sales increased by 10% successfully.")
    print(increased_sales)

# Run the test
test_increase_sales_by_percentage()
```
