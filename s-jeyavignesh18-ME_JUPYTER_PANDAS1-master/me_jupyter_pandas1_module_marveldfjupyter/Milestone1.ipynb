```python
import pandas as pd

def create_marvel_dataframe():
    """Create a Marvel DataFrame from a list of lists.

    Returns:
        pd.DataFrame: DataFrame containing Marvel character data.
    """
    marvel_data = [
        ['Spider-Man', 'male', 1962],
        ['Captain America', 'male', 1941],
        ['Iron Man', 'male', 1963],
        ['Thor', 'male', 1963],
        ['Hulk', 'male', 1962],
        ['Hawkeye', 'male', 1964],
        ['Doctor Strange', 'male', 1963],
        ['Scarlet Witch', 'female', 1964],
        ['Black Widow', 'female', 1964],
        ['Vision', 'male', 1968]
    ]
    # Create DataFrame
    df = pd.DataFrame(marvel_data, columns=['Name', 'Gender', 'Debut Year'])
    return df
print(create_marvel_dataframe())
```

```python
def test_create_marvel_dataframe():
    """Test function for create_marvel_dataframe to verify correct DataFrame creation."""
    df = create_marvel_dataframe()

    # Ensure DataFrame is not empty
    assert not df.empty, "Marvel DataFrame creation failed (empty DataFrame)"

    # Ensure shape is correct (10 rows, 3 columns)
    expected_shape = (10, 3)
    assert df.shape == expected_shape, f"Shape mismatch: Expected {expected_shape}, but found {df.shape}"

    print(f"✅ Test Passed: Marvel DataFrame created successfully with shape {df.shape}.")
    print(df)

# Run the test function
test_create_marvel_dataframe()
```
