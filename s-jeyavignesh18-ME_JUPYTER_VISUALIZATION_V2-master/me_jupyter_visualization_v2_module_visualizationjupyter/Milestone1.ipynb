```python
# TODO 2.1
# A company tracks sales of three products over time.
# Data: Monthly sales of three different products.
# months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
# product_A = [10000, 12000, 13000, 12500, 14000, 15000]
# product_B = [8000, 9500, 10500, 10200, 11000, 12000]
# product_C = [6000, 7200, 8000, 8200, 8800, 9500]

# Create line plots for above in same chart.
# Add title , label and show grids. 

import matplotlib.pyplot as plt

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
product_A = [10000, 12000, 13000, 12500, 14000, 15000]
product_B = [8000, 9500, 10500, 10200, 11000, 12000]
product_C = [6000, 7200, 8000, 8200, 8800, 9500]

# Create the plot
plt.figure()

# Plot each product with a different style for clarity
plt.plot(months, product_A, marker='o', linestyle='-', color='red', label='Product A')
plt.plot(months, product_B, marker='s', linestyle='--', color='green', label='Product B')
plt.plot(months, product_C, marker='^', linestyle='-.', color='blue', label='Product C')

# Add title and axis labels
plt.title("Monthly Sales Comparison - v2")
plt.xlabel("Months")
plt.ylabel("Sales ($)")

# Add gridlines
plt.grid(True)

# Add legend to distinguish the products
plt.legend()

# Display the plot
plt.show()
```

```python
# TODO 2.2
# For the chart created on 3 product sales data:
# Add labels to the plot , rename title as `Monthly Sales Comparison - V2`
# Show legends
# Modify colors and markers as well and remove grids
import matplotlib.pyplot as plt

# Monthly sales data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
product_A = [10000, 12000, 13000, 12500, 14000, 15000]
product_B = [8000, 9500, 10500, 10200, 11000, 12000]
product_C = [6000, 7200, 8000, 8200, 8800, 9500]

# Plot each product with distinct color and marker
plt.figure()
plt.plot(months, product_A, color='navy', marker='o', linestyle='-', linewidth=2, label='Product A')
plt.plot(months, product_B, color='crimson', marker='s', linestyle='--', linewidth=2, label='Product B')
plt.plot(months, product_C, color='darkorange', marker='^', linestyle='-.', linewidth=2, label='Product C')

# ✅ Add title and axis labels
plt.title("Monthly Sales Comparison - V2")
plt.xlabel("Month")
plt.ylabel("Sales ($)")

# ✅ Add legend
plt.legend()

# ✅ Remove the grid (make sure no grid is shown)
plt.grid(False)

# Display the plot
plt.show()
```

```python
# TODO 3
# To create bar chart use: plt.bar(<x-axis>,<y-axis>)
# For the product sale data create a bar chart with category on x-axis and sales on y-axis
# Add title and axis labels to your chart
# Data for reference:
# categories = ["Electronics", "Clothing", "Furniture", "Toys", "Books"]
# sales = [50000, 30000, 40000, 20000, 10000]  # Sales in dollars
import matplotlib.pyplot as plt

# Data
categories = ["Electronics", "Clothing", "Furniture", "Toys", "Books"]
sales = [50000, 30000, 40000, 20000, 10000]

# Create the bar plot
plt.figure()
plt.bar(categories, sales, color='skyblue')

# ✅ Add title and axis labels
plt.title("Product Sales by Category")
plt.xlabel("Product Category")
plt.ylabel("Sales ($)")

# Display the plot
plt.show()
```

```python
# TODO 4
# Create Histogram to analyze distribution of student scores
# Data:
# scores = np.random.randint(40, 100, 50)  #50 students with scores between 40-100
# Add titles and labels
# Add array for bin edges and edgecolor, also add gridlines from y-axis alone
import matplotlib.pyplot as plt
import numpy as np

# Data
scores = np.random.randint(40, 100, 50)  # 50 students with scores between 40 and 100

# Create the histogram
plt.figure()
bin_edges = np.arange(40, 101, 10)  # Bins: 40-50, 50-60, ..., 90-100

plt.hist(scores, bins=bin_edges, edgecolor='black', color='lightgreen')
plt.title("Distribution of Student Exam Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")

# Add gridlines only for y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# TODO 5
# Scatter plot syntax: plt.scatter(x-axis,y-axis)
# A real estate analyst examines whether larger houses cost more.
# Data: House sizes (square feet) and corresponding prices.
# house_size = np.random.randint(800, 3500, 50)  # House sizes in square feet
# house_price = house_size * np.random.uniform(100, 300) # Prices based on size
# Create plots and add titles and labels.

import matplotlib.pyplot as plt
import numpy as np

# Data
house_size = np.random.randint(800, 3500, 50)  # House sizes in square feet
house_price = house_size * np.random.uniform(100, 300)  # Prices proportional to size

# Create scatter plot
plt.figure()
plt.scatter(house_size, house_price, color='teal', alpha=0.7)
plt.title("House Size vs. Price")
plt.xlabel("House Size (sq ft)")
plt.ylabel("Price ($)")
plt.grid(True)  # Optionally add grid for better readability
plt.show()
```
