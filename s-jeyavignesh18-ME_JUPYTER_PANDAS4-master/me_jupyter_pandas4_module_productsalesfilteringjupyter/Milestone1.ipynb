# Product Sales Filtering and Sorting
This notebook processes the product sales dataset by filtering, sorting, and extracting top-performing products.

## Load the Dataset

```python
import pandas as pd

file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv'
df = pd.read_csv(file_path)

print(df.head())
print(df.info())
```

## Implement the Filtering and Sorting Function

```python
import pandas as pd

def filter_and_sort_products(df: pd.DataFrame) -> pd.DataFrame:
    """
    Filters products with sales ≥ 1000, selects only electronics,
    sorts by Sales descending, and returns top 100 products.

    Parameters:
        df (pd.DataFrame): DataFrame containing 'Product', 'Category', and 'Sales' columns.

    Returns:
        pd.DataFrame: Filtered and sorted DataFrame of top electronics products.
    """
    # Drop rows where Sales or Category is missing (so filtering is safe)
    df_clean = df.dropna(subset=["Sales", "Category"])

    # Ensure Sales is numeric (if it's read as string)
    df_clean["Sales"] = pd.to_numeric(df_clean["Sales"], errors="coerce")

    # Drop any rows where Sales conversion failed (NaN)
    df_clean = df_clean.dropna(subset=["Sales"])

    # Filter: sales at least 1000 and category "Electronics"
    filtered = df_clean[(df_clean["Sales"] >= 1000) & (df_clean["Category"] == "Electronics")]

    # Sort descending by sales
    sorted_df = filtered.sort_values(by="Sales", ascending=False)

    # Return top 100 (or fewer if not enough rows)
    top_100 = sorted_df.head(100).reset_index(drop=True)
    return top_100
```

## Apply the Filtering and Sorting Function and Display Results

```python
import pandas as pd

def filter_and_sort_products(df: pd.DataFrame) -> pd.DataFrame:
    """
    Filters out products with sales below 1000, keeps only electronics,
    sorts descending by 'Sales', and returns the top 100 items.
    """
    # Drop rows where Sales or Category is missing (so filtering is safe)
    df_clean = df.dropna(subset=["Sales", "Category"])

    # Convert Sales to numeric (in case it's string) — invalid parsing becomes NaN
    df_clean["Sales"] = pd.to_numeric(df_clean["Sales"], errors="coerce")
    # Drop rows where conversion failed
    df_clean = df_clean.dropna(subset=["Sales"])

    # Filter: retain only rows with Sales ≥ 1000 and Category = "Electronics"
    filtered = df_clean[(df_clean["Sales"] >= 1000) & (df_clean["Category"] == "Electronics")]

    # Sort descending by Sales
    sorted_filtered = filtered.sort_values(by="Sales", ascending=False)

    # Keep top 100
    top_100 = sorted_filtered.head(100).reset_index(drop=True)
    return top_100


# After implementing the function, you can run:
df_filtered_sorted = filter_and_sort_products(df)
df_filtered_sorted.head()
```

## Run Tests

```python
# Test: Checking filtering and sorting output
test_data = pd.DataFrame({
    'Product': ['Laptop', 'Phone', 'Shoes', 'TV', 'Tablet', 'Watch'],
    'Category': ['Electronics', 'Electronics', 'Fashion', 'Electronics', 'Electronics', 'Electronics'],
    'Sales': [1500, 900, 2000, 1200, 800, 3000]
})

expected_filtered_sorted = pd.DataFrame({
    'Product': ['Watch', 'Laptop', 'TV'],
    'Category': ['Electronics', 'Electronics', 'Electronics'],
    'Sales': [3000, 1500, 1200]
})

try:
    df_result = filter_and_sort_products(test_data)
    assert df_result.reset_index(drop=True).equals(expected_filtered_sorted), f"Expected:\n{expected_filtered_sorted}\nGot:\n{df_result}"
    print("✅ Test Passed: Filtering and sorting is correct.")
except Exception as e:
    print(f"❌ Test Failed: {e}")
```
