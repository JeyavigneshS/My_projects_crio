# 5-Star Business Analysis
This notebook processes the business rating dataset by filtering, grouping, and ranking cities based on 5-star businesses.

## Load the Dataset

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/5star_rating_business.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()  # TODO: Ensure dataset loads correctly
```

## Implement the Business Analysis Function

```python
import pandas as pd

def analyze_5star_businesses(df):
    """
    Analyzes the number of 5-star businesses per city and ranks cities.

    Parameters:
    df (pd.DataFrame): DataFrame containing 'Business', 'City', and 'Rating' columns.

    Returns:
    pd.DataFrame: Ranked cities based on count of 5-star businesses.
    """
    # Filter for 5-star businesses
    df_5star = df[df['Rating'] == 5]

    # Group by City and count the number of 5-star businesses
    city_counts = df_5star.groupby('City').size().reset_index(name='Count of 5-Star Businesses')

    # Sort cities by the count in descending order
    ranked_cities = city_counts.sort_values(by='Count of 5-Star Businesses', ascending=False)

    return ranked_cities
analyze_5star_businesses(df)
```

## Apply the Analysis Function and Display Results

```python
# Apply analysis function
df_ranked_cities = analyze_5star_businesses(df)  # TODO: Ensure function implementation

# Display the ranked cities DataFrame
df_ranked_cities.head()  # TODO: Verify ranked results
```

## Run Tests

```python
# Test: Checking 5-star business analysis output
test_data = pd.DataFrame({
    'Business': ['Cafe A', 'Restaurant B', 'Gym C', 'Salon D', 'Shop E', 'Cafe F', 'Bakery G'],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles', 'New York', 'Chicago'],
    'Rating': [5, 4, 5, 5, 5, 5, 3]
})

expected_ranked_cities = pd.DataFrame({
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Count of 5-Star Businesses': [3, 1, 1]
})

try:
    df_result = analyze_5star_businesses(test_data)

    # Sort both DataFrames by City before comparing
    df_result_sorted = df_result.sort_values(by='City').reset_index(drop=True)
    expected_sorted = expected_ranked_cities.sort_values(by='City').reset_index(drop=True)

    assert df_result_sorted.equals(expected_sorted), f"Expected:{expected_sorted} Got:{df_result_sorted}"
    print("**✅ Test Passed:** 5-star business analysis is correct.")
except Exception as e:
    print(f"**❌ Test Failed:** {e}")
```
