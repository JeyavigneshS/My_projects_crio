## Implement the Analysis Function

```python
import pandas as pd

def analyze_category_distribution(df):
    """
    Analyzes the distribution of product categories in the dataset.

    Parameters:
    df (pd.DataFrame): DataFrame containing 'Product', 'Category', and 'Sales' columns.

    Returns:
    tuple: A tuple containing:
           - The number of unique categories (int).
           - A dictionary with category names as keys and their respective product counts as values.
    """
    # Count the number of unique categories
    num_categories = df['Category'].nunique()

    # Count the number of products per category
    category_counts = df['Category'].value_counts().to_dict()

    return num_categories, category_counts

# Sample usage
url = "https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv"
df = pd.read_csv(url)

num_categories, category_counts = analyze_category_distribution(df)
print(f"Number of unique categories: {num_categories}")
print("Product counts per category:", category_counts)
```

## Apply the Analysis Function and Display Results

```python
# Apply analysis function
num_categories, category_counts = analyze_category_distribution(df)  # TODO: Ensure function implementation

# Display results
print(f"Number of Unique Categories: {num_categories}")  # TODO: Verify output
print(f"Category Distribution: {category_counts}")  # TODO: Ensure correct category distribution
```
