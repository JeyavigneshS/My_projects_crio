# Employee Salary Filtering
This notebook processes the employee dataset by filtering employees in the Marketing department earning more than $50,000.

## Load the Dataset

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/employee_data.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()  # TODO: Ensure dataset loads correctly
```

## Implement the Employee Filtering Function

```python
import pandas as pd

def filter_marketing_high_salary(df):
    """
    Filters employees in the 'Marketing' department earning more than $50,000.

    Parameters:
    df (pd.DataFrame): DataFrame containing 'Employee', 'Department', and 'Salary' columns.

    Returns:
    pd.DataFrame: Filtered DataFrame containing only qualifying employees.
    """
    # Filter for 'Marketing' department and salary greater than $50,000
    filtered_df = df.loc[(df['Department'] == 'Marketing') & (df['Salary'] > 50000)]

    return filtered_df
filter_marketing_high_salary(df)
```

## Apply the Filtering Function and Display Results

```python
# Apply filtering function
df_filtered = filter_marketing_high_salary(df)  # TODO: Ensure function implementation

# Display the filtered DataFrame
df_filtered.head()  # TODO: Verify filtered results
```

## Run Tests

```python
# Test: Checking filtered employee output
test_data = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['Marketing', 'Marketing', 'IT', 'Marketing', 'HR'],
    'Salary': [60000, 45000, 70000, 55000, 48000]
})

expected_filtered = pd.DataFrame({
    'Employee': ['Alice', 'David'],
    'Department': ['Marketing', 'Marketing'],
    'Salary': [60000, 55000]
})

try:
    df_result = filter_marketing_high_salary(test_data)
    assert df_result.reset_index(drop=True).equals(expected_filtered), f"Expected:\n{expected_filtered}\nGot:\n{df_result}"
    print("**✅ Test Passed:** Employee filtering is correct.")
except Exception as e:
    print(f"**❌ Test Failed:** {e}")
```
