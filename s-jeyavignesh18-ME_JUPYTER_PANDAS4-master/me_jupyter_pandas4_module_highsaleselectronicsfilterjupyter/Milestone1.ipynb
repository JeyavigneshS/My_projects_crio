# Product Sales Filtering
This notebook processes the product sales dataset by filtering products based on sales and category criteria.

## Load the Dataset

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()  # TODO: Ensure dataset loads correctly
```

## Implement the Filtering Function

```python
import pandas as pd

def filter_products(df):
    """
    Filters products with sales greater than $1,000 and belonging to the 'Electronics' category.

    Parameters:
    df (pd.DataFrame): DataFrame containing 'Product', 'Category', and 'Sales' columns.

    Returns:
    pd.DataFrame: Filtered DataFrame.
    """
    # Ensure the 'Sales' column is numeric
    df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

    # Filter products with sales > 1000 and category 'Electronics'
    filtered_df = df.loc[(df['Sales'] > 1000) & (df['Category'] == 'Electronics')]

    return filtered_df
filter_products(df)
```

## Apply the Filtering Function and Display Results

```python
# Apply filtering function
df_filtered = filter_products(df)  # TODO: Ensure function implementation

# Display the filtered DataFrame
df_filtered.head()  # TODO: Verify filtered results
```

## Run Tests

```python
# Test: Checking filtered products output
test_data = pd.DataFrame({
    'Product': ['Laptop', 'Phone', 'Shoes', 'TV', 'Table'],
    'Category': ['Electronics', 'Electronics', 'Fashion', 'Electronics', 'Furniture'],
    'Sales': [1500, 900, 2000, 1200, 800]
})

expected_filtered = pd.DataFrame({
    'Product': ['Laptop', 'TV'],
    'Category': ['Electronics', 'Electronics'],
    'Sales': [1500, 1200]
})

try:
    df_result = filter_products(test_data)
    assert df_result.reset_index(drop=True).equals(expected_filtered), f"Expected:\n{expected_filtered}\nGot:\n{df_result}"
    print("**✅ Test Passed:** Product filtering is correct.")
except Exception as e:
    print(f"**❌ Test Failed:** {e}")
```
