# Analyzing Product Sales Data
This notebook helps in analyzing product sales by category to understand revenue distribution, average sales per product, and the number of products sold in each category.

You will implement a function to:
- Group products by category.
- Compute total sales, average sales per product, and number of products in each category.
- Return the results as a dictionary.

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

## Implement the Function
Complete the `analyze_sales` function to compute:
- Total sales per category using `sum()`.
- Average sales per product using `mean()`.
- Number of products in each category using `count()`.

```python
import pandas as pd

def analyze_sales(dataframe: pd.DataFrame) -> dict:
    """
    Groups the dataframe by 'Category' and computes total sales, average sales, and product counts.

    Parameters:
        dataframe (pd.DataFrame): DataFrame containing 'Product', 'Category', and 'Sales' columns.

    Returns:
        dict: A dictionary mapping each Category to a dict with keys 'sum', 'mean', 'count'.
    """
    # Ensure Sales column is numeric (in case it's string)
    dataframe = dataframe.copy()
    dataframe['Sales'] = pd.to_numeric(dataframe['Sales'], errors='coerce')
    
    # Drop rows where Sales or Category is missing (so metrics make sense)
    dataframe = dataframe.dropna(subset=['Sales', 'Category'])
    
    # Group by category
    grouped = dataframe.groupby('Category')
    
    # Compute sum, mean, count
    sum_series = grouped['Sales'].sum()
    mean_series = grouped['Sales'].mean()
    count_series = grouped['Sales'].count()
    
    # Build the output dictionary
    output = {}
    for category in sum_series.index:
        output[category] = {
            'sum': sum_series[category],
            'mean': mean_series[category],
            'count': int(count_series[category])  # count is integer
        }
    
    return output
```

## Run the Test
Run the test below after implementing the function to validate your solution.

```python
# Test: Checking category-wise sales analysis
test_data = pd.DataFrame({
    'Product': ['TV', 'Blender', 'Smartphone', 'Book'],
    'Category': ['Furniture', 'Home', 'Furniture', 'Electronics'],
    'Sales': [2079, 3551, 2158, 3948]
})

# Expected Output
expected_output = {
    'Furniture': {'sum': 4237, 'mean': 2118.5, 'count': 2},
    'Home': {'sum': 3551, 'mean': 3551.0, 'count': 1},
    'Electronics': {'sum': 3948, 'mean': 3948.0, 'count': 1}
}

result = analyze_sales(test_data)

try:
    assert result == expected_output
    print("✅ Test Passed: Sales analysis is correct.")
except AssertionError:
    print("❌ Test Failed: Output does not match expected result.")
```
