# Analyzing Canceled Shipments
This notebook helps analyze shipment cancellations within a fixed period (January to March 2024) to identify cancellation trends.

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/shipment_data.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

## Implement the Function
Complete the `analyze_canceled_shipments` function to compute:
- The number of canceled shipments per month for **January, February, and March 2024**.
- Ensure the output is a dictionary.

```python
def analyze_canceled_shipments(dataframe):
    # Ensure 'shipment_date' is in datetime format
    dataframe['shipment_date'] = pd.to_datetime(dataframe['shipment_date'], errors='coerce')
    
    # Filter for status 'Cancelled'
    df = dataframe[dataframe['status'] == 'Cancelled']
    
    # Filter for shipment_date in Jan, Feb, Mar 2024
    # (Month is in [1,2,3] and year == 2024)
    df = df[(df['shipment_date'].dt.year == 2024) & (df['shipment_date'].dt.month.isin([1,2,3]))]
    
    # Convert shipment_date to 'YYYY-MM' string WITHOUT using strftime
    df['year_month'] = df['shipment_date'].dt.year.astype(str) + '-' + df['shipment_date'].dt.month.astype(str).str.zfill(2)
    
    # Count canceled shipments per month
    counts = df.groupby('year_month').size()
    
    # Return as dictionary
    return counts.to_dict()
```

## Run the Test
Run the test below after implementing the function to validate your solution.

```python
# Test: Checking shipment cancellation analysis
test_data = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'shipment_date': ['2024-01-10', '2024-02-15', '2024-03-01', '2024-03-08', '2024-04-06', '2024-04-07'],
    'status': ['Cancelled', 'Cancelled', 'Cancelled', 'Cancelled', 'Cancelled', 'Cancelled']
})

# Expected Output
expected_output = {
    '2024-01': 1,
    '2024-02': 1,
    '2024-03': 2
}

result = analyze_canceled_shipments(test_data)

try:
    assert result == expected_output
    print("✅ Test Passed: Canceled shipment analysis is correct.")
except AssertionError:
    print("❌ Test Failed: Output does not match expected result.")
```
