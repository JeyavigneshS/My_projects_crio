# Product Sales Data Analysis
This notebook processes the product sales dataset by cleaning duplicates and analyzing product categories.

## Load the Dataset

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()  # TODO: Ensure dataset loads correctly
```

## Implement the Analysis Function

```python
import pandas as pd

def analyze_product_categories(df):
    """
    Analyzes product categories after removing duplicates.

    Steps:
    1. Drop duplicate rows (so same product isn’t counted twice).
    2. Count unique categories (nunique).
    3. List unique categories (unique → tolist).
    4. Count number of products per category (value_counts → to_dict).

    Parameters:
        df (pd.DataFrame): DataFrame with columns 'Product', 'Category', 'Sales'.

    Returns:
        tuple: (num_unique_categories, list_of_unique_categories, dict_of_counts_per_category)
    """
    # Drop duplicate rows (keeping first occurrence)
    df_clean = df.drop_duplicates()

    # Count unique categories
    num_unique = df_clean['Category'].nunique()

    # List unique categories
    unique_cats = df_clean['Category'].unique().tolist()

    # Count products in each category
    counts = df_clean['Category'].value_counts().to_dict()

    return num_unique, unique_cats, counts
analyze_product_categories(df)
```

## Apply the Analysis Function and Display Results

```python
# Apply analysis function
num_categories, unique_categories, category_counts = analyze_product_categories(df)  # TODO: Ensure function implementation

# Display results
print(f"Number of Unique Categories: {num_categories}")  # TODO: Verify output
print(f"Unique Categories: {unique_categories}")  # TODO: Check category listing
print(f"Product Count per Category: {category_counts}")  # TODO: Ensure correct category counts
```

## Run Tests

```python
# Test: Checking category analysis output after removing duplicates
test_data = pd.DataFrame({
    'Product': ['Laptop', 'Phone', 'Shoes', 'Laptop', 'TV', 'Phone'],
    'Category': ['Electronics', 'Electronics', 'Fashion', 'Electronics', 'Electronics', 'Electronics'],
    'Sales': [1500, 900, 2000, 1500, 1200, 900]
})

expected_unique_categories = ['Electronics', 'Fashion']
expected_category_counts = {'Electronics': 3, 'Fashion': 1}  # Duplicates removed

try:
    num_categories, unique_categories, category_counts = analyze_product_categories(test_data)
    assert set(unique_categories) == set(expected_unique_categories), f"Expected {expected_unique_categories}, but got {unique_categories}"
    assert category_counts == expected_category_counts, f"Expected {expected_category_counts}, but got {category_counts}"
    print("**✅ Test Passed:** Product category analysis is correct.")
except Exception as e:
    print(f"**❌ Test Failed:** {e}")
```
