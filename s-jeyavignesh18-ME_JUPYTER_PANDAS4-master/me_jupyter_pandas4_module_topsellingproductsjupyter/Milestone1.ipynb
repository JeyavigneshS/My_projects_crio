# Product Sales Ranking
This notebook processes the product sales dataset by ranking products based on their sales figures and selecting the top N products.

## Load the Dataset

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()  # TODO: Ensure dataset loads correctly
```

## Implement the Product Ranking Function

```python
import pandas as pd

def rank_top_products(df, top_n=3):
    """
    Rank products based on their sales figures and select the top N products.

    Parameters:
    df (pd.DataFrame): DataFrame containing 'Product', 'Category', and 'Sales' columns.
    top_n (int): Number of top products to return.

    Returns:
    pd.DataFrame: Ranked DataFrame containing top N products.
    """
    # Ensure 'Sales' is numeric
    df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

    # Rank products based on 'Sales' in descending order
    df['Rank'] = df['Sales'].rank(ascending=False, method='min')

    # Filter top N products
    top_products = df[df['Rank'] <= top_n].sort_values(by='Rank')

    # Return the top N products with 'Product', 'Sales', and 'Rank'
    return top_products[['Product', 'Sales', 'Rank']]

# Sample usage
url = "https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv"
df = pd.read_csv(url)
top_n = 3
top_products = rank_top_products(df, top_n)
print(top_products)
```

## Apply the Ranking Function and Display Results

```python
# Apply ranking function
df_top_products = rank_top_products(df, top_n=3)  # TODO: Ensure function implementation

# Display the top ranked products DataFrame
df_top_products.head()  # TODO: Verify ranking results
```
