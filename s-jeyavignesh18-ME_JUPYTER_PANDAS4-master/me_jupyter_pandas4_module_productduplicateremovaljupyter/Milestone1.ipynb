# Cleaning Product Sales Data
This notebook helps in cleaning product sales data by removing duplicate product entries.

You will implement a function to remove duplicate product entries while keeping the first occurrence.

## Load the Dataset
The dataset is stored in a CSV file and can be loaded directly using Pandas.

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

## Implement the Function
Complete the `clean_product_data` function to remove duplicate product entries based on their names.

### Steps:
- Identify duplicate rows based on the 'Product' column.
- Remove duplicate product names while keeping only the first occurrence.
- Return the cleaned dataset.

```python
import pandas as pd

def clean_product_data(df):
    """
    Cleans product sales data by removing duplicate product names (keeping first occurrence).

    Parameters:
        df (pd.DataFrame): DataFrame containing 'Product', 'Category', and 'Sales' columns.

    Returns:
        pd.DataFrame: Cleaned DataFrame without duplicate product names.
    """
    # Print before cleaning
    print("Before cleaning (first few rows):")
    print(df.head(10))
    print("Total rows before cleaning:", len(df))

    # Drop duplicates based on Product column, keeping first occurrence
    df_clean = df.drop_duplicates(subset=['Product'], keep='first')

    # Print after cleaning
    print("\nAfter cleaning (first few rows):")
    print(df_clean.head(10))
    print("Total rows after cleaning:", len(df_clean))

    return df_clean
data = [
    {"Product": "TV", "Category": "Furniture", "Sales": 2079},
    {"Product": "Blender", "Category": "Home", "Sales": 3551},
    {"Product": "Smartphone", "Category": "Furniture", "Sales": 2158},
    {"Product": "TV", "Category": "Furniture", "Sales": 2079},  # Duplicate
    {"Product": "Book", "Category": "Electronics", "Sales": 3948}
]
df = pd.DataFrame(data)

cleaned = clean_product_data(df)
```

## Run the Test
Run the test below after implementing the function to validate your solution.

```python
# Test: Checking duplicate removal
test_data = pd.DataFrame({
    'Product': ['TV', 'Blender', 'Smartphone', 'TV', 'Book'],
    'Category': ['Furniture', 'Home', 'Furniture', 'Furniture', 'Electronics'],
    'Sales': [2079, 3551, 2158, 2079, 3948]
})

# Expected cleaned dataset
expected_data = pd.DataFrame({
    'Product': ['TV', 'Blender', 'Smartphone', 'Book'],
    'Category': ['Furniture', 'Home', 'Furniture', 'Electronics'],
    'Sales': [2079, 3551, 2158, 3948]
})

result_data = clean_product_data(test_data)

try:
    pd.testing.assert_frame_equal(result_data.reset_index(drop=True),
                                  expected_data.reset_index(drop=True))
    print("✅ Test Passed: Duplicate product names removed successfully.")
except AssertionError as e:
    print(f"❌ Test Failed:\n{e}")
```
