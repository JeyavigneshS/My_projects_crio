# Student Score Ranking
This notebook implements the ranking of students based on their total scores using different ranking methods in Pandas.

## Load the Dataset

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/student_scores.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()  # TODO: Ensure dataset loads correctly
```

## Implement the Ranking Function

```python
import pandas as pd

def rank_students(df: pd.DataFrame) -> pd.DataFrame:
    """
    Given a DataFrame with columns 'Student', 'Math', 'Science', 'English',
    this function computes the total score and then ranks students using
    four different ranking methods (avg, min, max, dense), in descending
    order (higher total → better rank).

    Parameters:
        df (pd.DataFrame): DataFrame containing student scores.

    Returns:
        pd.DataFrame: The same DataFrame augmented with columns:
            'Total Score', 'Rank_Avg', 'Rank_Min', 'Rank_Max', 'Rank_Dense'
    """
    # Make a copy so we don’t mutate original input
    df2 = df.copy()

    # Convert subject columns to numeric (if not already) — errors become NaN
    for subj in ["Math", "Science", "English"]:
        df2[subj] = pd.to_numeric(df2[subj], errors="coerce")

    # Compute total score (sum across the three subjects) — axis = 1
    df2["Total Score"] = df2[["Math", "Science", "English"]].sum(axis=1)

    # Now perform ranking on the Total Score, in descending order
    df2["Rank_Avg"] = df2["Total Score"].rank(method="average", ascending=False)
    df2["Rank_Min"] = df2["Total Score"].rank(method="min", ascending=False)
    df2["Rank_Max"] = df2["Total Score"].rank(method="max", ascending=False)
    df2["Rank_Dense"] = df2["Total Score"].rank(method="dense", ascending=False)

    return df2
```

## Apply the Ranking Function and Display Results

```python
# Apply ranking function
df_ranked = rank_students(df)  # TODO: Ensure function implementation

# Display the ranked DataFrame
df_ranked.head()  # TODO: Verify ranking results
```

## Run Tests

```python
# Test: Checking student ranking output
test_data = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math': [90, 85, 78, 92, 88],
    'Science': [85, 80, 75, 95, 82],
    'English': [88, 86, 80, 90, 84]
})

expected_ranked = pd.DataFrame({
    'Student': ['David', 'Alice', 'Eve', 'Bob', 'Charlie'],
    'Total Score': [277, 263, 254, 251, 233],
    'Rank_Avg': [1, 2, 3, 4, 5],
    'Rank_Min': [1, 2, 3, 4, 5],
    'Rank_Max': [1, 2, 3, 4, 5],
    'Rank_Dense': [1, 2, 3, 4, 5]
})

try:
    df_result = rank_students(test_data)
    df_result = df_result[['Student', 'Total Score', 'Rank_Avg', 'Rank_Min', 'Rank_Max', 'Rank_Dense']]

    # Convert rank columns to integer type to match expected output
    rank_columns = ['Rank_Avg', 'Rank_Min', 'Rank_Max', 'Rank_Dense']
    df_result[rank_columns] = df_result[rank_columns].astype(int)

    # Sort both expected and result DataFrames before comparison
    df_result = df_result.sort_values(by="Total Score", ascending=False).reset_index(drop=True)
    expected_ranked = expected_ranked.sort_values(by="Total Score", ascending=False).reset_index(drop=True)

    assert df_result.equals(expected_ranked), \
        f"Expected:\n{expected_ranked}\nGot:\n{df_result}"

    print("✅ **Test Passed:** Student ranking function is correct.")
except Exception as e:
    print(f"❌ **Test Failed:** {e}")
```
