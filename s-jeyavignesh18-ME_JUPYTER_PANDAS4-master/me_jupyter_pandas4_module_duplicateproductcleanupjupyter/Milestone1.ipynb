# Cleaning Product Sales Data
This notebook helps in cleaning product sales data by:
- Removing duplicate product names.
- Filtering out low-sales products based on a given threshold.

```python
import pandas as pd

# Load the dataset
file_path = 'https://gitlab.crio.do/public_content/da-ds-datasets/pyt-102-datasets/-/raw/main/product_sales_dataset.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

## Implement the Function
Complete the `clean_product_data` function to:
- Remove duplicate products (keeping the first occurrence).
- Remove products with sales below a given threshold.

```python
import pandas as pd

def clean_product_data(data, sales_threshold):
    """
    Clean product sales data by removing duplicate products and filtering by sales threshold.

    Parameters:
    data (pd.DataFrame): DataFrame containing 'Product', 'Category', and 'Sales' columns.
    sales_threshold (int): Minimum sales value required to retain a product.

    Returns:
    pd.DataFrame: Cleaned DataFrame without duplicate product names and low-sales products.
    """
    # Remove duplicate products, keeping the first occurrence
    data_cleaned = data.drop_duplicates(subset=['Product'], keep='first')

    # Filter products with sales greater than or equal to the sales_threshold
    data_cleaned = data_cleaned[data_cleaned['Sales'] >= sales_threshold]

    return data_cleaned
```

## Run the Test
Run the test below after implementing the function to validate your solution.

```python
# Test: Checking duplicate removal and low-sales filtering
test_data = pd.DataFrame({
    'Product': ['TV', 'Blender', 'Smartphone', 'TV', 'Book'],
    'Category': ['Furniture', 'Home', 'Furniture', 'Furniture', 'Electronics'],
    'Sales': [2079, 800, 2158, 2079, 500]
})

# Expected cleaned dataset (Threshold = 1000)
expected_data = pd.DataFrame({
    'Product': ['TV', 'Smartphone'],
    'Category': ['Furniture', 'Furniture'],
    'Sales': [2079, 2158]
})

result_data = clean_product_data(test_data, 1000)

try:
    pd.testing.assert_frame_equal(result_data.reset_index(drop=True), expected_data.reset_index(drop=True))
    print("✅ Test Passed: Duplicate removal and sales filtering working correctly.")
except AssertionError as e:
    print(f"❌ Test Failed:\n{e}")
```
