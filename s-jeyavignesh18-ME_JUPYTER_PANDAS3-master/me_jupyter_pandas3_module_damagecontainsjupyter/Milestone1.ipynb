```python
import pandas as pd

def extract_product_issues(df):
    """
    Extracts reviews mentioning product issues such as 'missing', 'damaged', or 'broken'.

    Parameters:
        df (pd.DataFrame): The input DataFrame containing customer reviews.

    Returns:
        pd.DataFrame: DataFrame containing reviews that mention product issues.
    """
    # Define the keywords to search for
    keywords = ['missing', 'damaged', 'broken']
    
    # Create a case-insensitive regex pattern to match any of the keywords
    pattern = '|'.join(keywords)
    
    # Filter the DataFrame to include only rows where the 'Customer Review' column contains any of the keywords
    filtered_df = df[df['Customer Review'].str.contains(pattern, case=False, na=False)]
    
    # Print the original and filtered DataFrames
    print("Original DataFrame:")
    print(df)
    print("\nFiltered DataFrame:")
    print(filtered_df)
    
    return filtered_df
# Sample DataFrame
data = {
    'Review ID': [101, 102, 103, 104, 105, 106],
    'Customer Review': [
        'The delivery was late, very disappointed!',
        'Great product, arrived on time.',
        'I love the phone but the charger was missing.',
        'Delivery took longer than expected, not happy.',
        'Excellent quality, highly recommend it!',
        'The package was damaged upon arrival.'
    ]
}

df = pd.DataFrame(data)

# Extract reviews mentioning product issues
result_df = extract_product_issues(df)


extract_product_issues(df)
```
