```python
import pandas as pd

def extract_delivery_issues(df):
    """
    Extracts reviews mentioning delivery-related issues.

    Parameters:
        df (pd.DataFrame): The input DataFrame containing customer reviews.

    Returns:
        pd.DataFrame: DataFrame containing reviews that mention delivery issues.
    """
    # Define a list of keywords related to delivery issues
    keywords = ['delivery', 'late', 'delay']

    # Create a case-insensitive regex pattern for the keywords
    pattern = '|'.join(keywords)

    # Filter rows where 'Customer Review' contains any of the keywords
    filtered_df = df[df['Customer Review'].str.contains(pattern, case=False, na=False, regex=True)]

    # Print the original and filtered DataFrames
    print("Original DataFrame:")
    print(df)
    print("\nFiltered DataFrame (Delivery Issues):")
    print(filtered_df)

    return filtered_df
extract_delivery_issues(df)
```
