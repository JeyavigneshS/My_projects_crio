```python
import pandas as pd

def detect_missing_values(df):
    """
    Detects missing values in a DataFrame.

    Parameters:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        tuple: A tuple containing two elements:
            - A DataFrame indicating missing values (True/False)
            - A Series with the count of missing values per column
    """
    # Detect missing values
    missing_values_df = df.isnull()
    missing_counts = df.isnull().sum()

    # Print the results
    print("Missing Values DataFrame:")
    print(missing_values_df)
    print("\nMissing Values Count per Column:")
    print(missing_counts)

    return missing_values_df, missing_counts
data = {
    'Customer': ['Alice', 'Bob', 'Charlie', 'David'],
    'Total_Spend': [200.50, None, 150.75, None]
}
df = pd.DataFrame(data)

# detect_missing_values(df)
```
