```python
import pandas as pd

def categorize_purchase(amount):
    """
    Categorizes purchase amounts into 'Low', 'Medium', or 'High'.

    Parameters:
        amount (int or float): The purchase amount.

    Returns:
        str: Category of the purchase amount.
    """
    if amount < 100:
        return "Low"
    elif 100 <= amount <= 200:
        return "Medium"
    else:
        return "High"
amount=50
categorize_purchase(amount)
```

```python

def apply_purchase_category(df):
    """
    Applies the categorize_purchase function to the 'Purchase Amount' column.

    Parameters:
        df (pd.DataFrame): The input DataFrame containing purchase amounts.

    Returns:
        pd.DataFrame: DataFrame with an added 'Purchase Category' column.
    """
    df['Purchase Category'] = df['Purchase Amount'].apply(categorize_purchase)
    print(df)
    return df



# Create sample DataFrame
df = pd.DataFrame({
    "Customer Name": ["Alice", "Bob", "Charlie", "David"],
    "Purchase Amount": [100, 250, 75, 150]
})

# Now call the function
print(apply_purchase_category(df))
```

```python
def test_apply_purchase_category():
    """Test if the 'Purchase Category' column is correctly added and categorized."""
    # Check if apply() is used in the function
    source_code = inspect.getsource(apply_purchase_category)
    assert "apply" in source_code, "apply() is not used in the function"

    # Create test DataFrame
    df = pd.DataFrame({
        "Customer Name": ["Alice", "Bob", "Charlie", "David"],
        "Purchase Amount": [100, 250, 75, 150]
    })

    # Apply function
    result = apply_purchase_category(df)

    # Validate DataFrame structure
    assert isinstance(result, pd.DataFrame), "Function did not return a DataFrame"
    assert "Purchase Category" in result.columns, "'Purchase Category' column was not added"

    # Validate category assignment
    expected_categories = ["Medium", "High", "Low", "Medium"]
    assert list(result["Purchase Category"]) == expected_categories, "Purchase categories were not assigned correctly"

    print("✅ Test Passed: apply_purchase_category() is implemented correctly.")

# Run the test
test_apply_purchase_category()
```
