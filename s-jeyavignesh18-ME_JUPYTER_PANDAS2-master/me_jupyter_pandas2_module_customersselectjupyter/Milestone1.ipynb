```python
import pandas as pd
import inspect
```

```python
def select_dataframe_values():
    """
    Creates a Pandas DataFrame with sample data and returns selected values using iloc.

    Returns:
        tuple: Selected values using iloc (first_row, first_three_rows, third_customer_amount, last_row)
    """
    data = {
        "Unnamed_Column_1": ["Alice", "Bob", "Charlie", "David"],
        "Unnamed_Column_2": [250, 100, 75, 150],
        "Unnamed_Column_3": ["New York", "Los Angeles", "Chicago", "Houston"]
    }
    df = pd.DataFrame(data)
    
    # Use iloc to select:
    first_row = df.iloc[0]               # first row
    first_three_rows = df.iloc[0:3]      # first three rows
    third_customer_amount = df.iloc[2, 1]  # 3rd row, 2nd column (Unnamed_Column_2)
    last_row = df.iloc[-1]               # last row
    
    return (first_row, first_three_rows, third_customer_amount, last_row)
select_dataframe_values()
```

```python
def test_select_dataframe_values():
    """Test if df.iloc is used in the function and returns expected values."""
    # Check if df.iloc is used in the function
    source_code = inspect.getsource(select_dataframe_values)
    assert "iloc" in source_code, "df.iloc is not used in the function"

    # Check if the function returns expected values
    results = select_dataframe_values()
    assert isinstance(results, tuple), "Function did not return a tuple of values"
    assert len(results) == 4, "Function did not return the expected number of values"

    print("✅ Test Passed: select_dataframe_values() is implemented correctly.")

test_select_dataframe_values()
```

```python
import pandas as pd
import inspect

def filter_customers_by_city():
    """
    Creates a Pandas DataFrame and filters customers by city using loc.

    Returns:
        pd.DataFrame: Filtered DataFrame with selected columns.
    """
    data = {
        "Customer Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
        "Purchase Amount": [250, 100, 75, 150, 300],
        "City": ["New York", "Los Angeles", "Chicago", "New York", "Houston"]
    }
    df = pd.DataFrame(data)
    # Use loc to filter rows where City == "New York" and select only the "Customer Name" and "Purchase Amount" columns
    filtered = df.loc[df["City"] == "New York", ["Customer Name", "Purchase Amount"]]
    return filtered
filter_customers_by_city()
```

```python
def test_filter_customers_by_city():
    """Test if df.loc is used in the function and returns a DataFrame."""
    # Check if df.loc is used in the function
    source_code = inspect.getsource(filter_customers_by_city)
    assert "loc" in source_code, "df.loc is not used in the function"

    # Check if the function returns a DataFrame
    result = filter_customers_by_city()
    assert isinstance(result, pd.DataFrame), "Function did not return a DataFrame"

    print("✅ Test Passed: filter_customers_by_city() is implemented correctly.")

test_filter_customers_by_city()
```
