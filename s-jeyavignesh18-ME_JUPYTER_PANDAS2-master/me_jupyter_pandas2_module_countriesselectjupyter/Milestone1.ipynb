```python
import pandas as pd
import inspect
```

```python
def select_country_data():
    """
    Creates a Pandas DataFrame with country-related data and returns selected values using iloc.

    Returns:
        tuple: Selected values using iloc.
    """
    data = {
        'Population': [35.467, 63.951, 80.94, 60.665, 127.061, 64.511, 318.523],
        'GDP': [1785387, 2833687, 3874437, 2167744, 4602367, 2950039, 17348075],
        'Surface Area': [9984670, 640679, 357114, 301336, 377930, 242495, 9525067],
        'HDI': [0.913, 0.888, 0.916, 0.873, 0.891, 0.907, 0.915],
        'Continent': ['America', 'Europe', 'Europe', 'Europe', 'Asia', 'Europe', 'America']
    }
    countries = [
        'Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'
    ]
    df = pd.DataFrame(data, index=countries)

    # 1. First country's data (0th index)
    first_country = df.iloc[0]
    print(first_country)

    # 2. Last country's data
    last_country = df.iloc[-1]
    print(last_country)

    # 3. Data of 0th, 1st, and last country together
    first_second_last = df.iloc[[0, 1, -1]]
    print(first_second_last)

    # 4. Data of 1st and 2nd country together
    second_third = df.iloc[[1, 2]]
    print(second_third)

    return (first_country, last_country, first_second_last, second_third)
print(select_country_data())
```

```python
def test_select_country_data():
    """Test if df.iloc is used in the function and returns expected length of values."""
    # Check if df.iloc is used in the function
    source_code = inspect.getsource(select_country_data)
    assert "iloc" in source_code, "df.iloc is not used in the function"

    # Check if the function returns expected number of values
    results = select_country_data()
    assert isinstance(results, tuple), "Function did not return a tuple"
    assert len(results) == 4, "Function did not return the expected number of values"

    print("✅ Test Passed: select_country_data() is implemented correctly.")

test_select_country_data()
```

```python
import pandas as pd

def filter_country_data():
    """
    Creates a Pandas DataFrame with country-related data and returns selected values using loc.

    Returns:
        pd.Series: Selected population values using loc.
    """
    data = {
        'Population': [35.467, 63.951, 80.94, 60.665, 127.061, 64.511, 318.523],
        'GDP': [1785387, 2833687, 3874437, 2167744, 4602367, 2950039, 17348075],
        'Surface Area': [9984670, 640679, 357114, 301336, 377930, 242495, 9525067],
        'HDI': [0.913, 0.888, 0.916, 0.873, 0.891, 0.907, 0.915],
        'Continent': ['America', 'Europe', 'Europe', 'Europe', 'Asia', 'Europe', 'America']
    }
    countries = [
        'Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'
    ]
    df = pd.DataFrame(data, index=countries)

    # Select populations from France to Italy (inclusive) using loc and squeeze to return Series
    population_series = df.loc['France':'Italy', 'Population']
    print(population_series)

    return population_series
print(filter_country_data())
```

```python
def test_filter_country_data():
    """Test if df.loc is used in the function and returns a Series."""
    # Check if df.loc is used in the function
    source_code = inspect.getsource(filter_country_data)
    assert "loc" in source_code, "df.loc is not used in the function"

    # Check if the function returns a Pandas Series
    result = filter_country_data()
    assert isinstance(result, pd.Series), "Function did not return a Pandas Series"

    print("✅ Test Passed: filter_country_data() is implemented correctly.")

test_filter_country_data()
```
