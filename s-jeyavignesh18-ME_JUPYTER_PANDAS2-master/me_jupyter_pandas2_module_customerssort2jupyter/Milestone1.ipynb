```python
import pandas as pd

def sort_dataframe(df):
    """
    Sorts the DataFrame in different ways:
    1. By Purchase Amount (highest to lowest)
    2. By Customer Name (alphabetical order)
    3. By City (A-Z) and then Purchase Amount (high to low)
    4. By Purchase Date (oldest to newest)

    Parameters:
        df (pd.DataFrame): The input DataFrame containing purchase data.

    Returns:
        tuple: Four sorted DataFrames.
    """
    # Sort by Purchase Amount (highest to lowest)
    sorted_amount = df.sort_values(by='Purchase Amount', ascending=False)
    print("Sorted by Purchase Amount:")
    print(sorted_amount)
    print()

    # Sort by Customer Name (alphabetical order)
    sorted_name = df.sort_values(by='Customer Name')
    print("Sorted by Customer Name:")
    print(sorted_name)
    print()

    # Sort by City (A-Z) and then Purchase Amount (high to low)
    sorted_city_amount = df.sort_values(by=['City', 'Purchase Amount'], ascending=[True, False])
    print("Sorted by City and Purchase Amount:")
    print(sorted_city_amount)
    print()

    # Sort by Purchase Date (oldest to newest)
    df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
    sorted_date = df.sort_values(by='Purchase Date')
    print("Sorted by Purchase Date:")
    print(sorted_date)
    print()

    return sorted_amount, sorted_name, sorted_city_amount, sorted_date

# Sample DataFrame
df = pd.DataFrame({
    "Customer Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace"],
    "Purchase Amount": [120, 250, 75, 300, 150, 200, 100],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "New York", "Chicago", "Los Angeles"],
    "Purchase Date": ["2024-03-10", "2024-02-20", "2024-01-15", "2024-03-05", "2024-02-28", "2024-01-10", "2024-02-25"]
})

# Apply the sorting function
sorted_amount, sorted_name, sorted_city_amount, sorted_date = sort_dataframe(df)
sort_dataframe(df)
```
