```python
import pandas as pd
import inspect
```

```python


def create_country_dataframe():
    """
    Creates a Pandas DataFrame with country-related data and prints its information.

    Returns:
        pd.DataFrame: The created DataFrame.
    """
    data = {
        'Population': [35.467, 63.951, 80.94, 60.665, 127.061, 64.511, 318.523],
        'GDP': [1785387, 2833687, 3874437, 2167744, 4602367, 2950039, 17348075],
        'Surface Area': [9984670, 640679, 357114, 301336, 377930, 242495, 9525067],
        'HDI': [0.913, 0.888, 0.916, 0.873, 0.891, 0.907, 0.915],
        'Continent': ['America', 'Europe', 'Europe', 'Europe', 'Asia', 'Europe', 'America']
    }
    countries = ["Canada", "France", "Germany", "Italy", "Japan", "United Kingdom", "United States"]
    df = pd.DataFrame(data, index=countries)
    df.info()
    return df
create_country_dataframe()
```

```python
def test_create_country_dataframe():
    """Test if df.info() is called and a DataFrame is returned."""
    # Check if df.info() is used in the function
    source_code = inspect.getsource(create_country_dataframe)
    assert ".info()" in source_code, "df.info() is not called in the function"

    # Check if the function returns a DataFrame
    df = create_country_dataframe()
    assert isinstance(df, pd.DataFrame), "Function did not return a DataFrame"

    print("✅ Test Passed: create_country_dataframe() is implemented correctly.")

test_create_country_dataframe()
```
